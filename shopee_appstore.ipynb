```python
import pandas as pd
import numpy as np
import json
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

from app_store_scraper import AppStore
data = AppStore(country='id', app_name='Shopee 11.11 Big Sale', app_id='959841443')
data.review(how_many=2000)
```

```python
data.review
```

```python
data1 = pd.DataFrame(np.array(data.reviews),columns=['review'])
```

```python
data2 = data1.join(pd.DataFrame(data1.pop('review').tolist()))
```

```python
data2 = data2.sort_values('date', ascending=False)
data2.head()
```

```python
Shopee = data2[['date','userName','review','rating']]
Shopee
```

```python
Shopee.to_csv('Shopee.csv', index=False)
```
